```python
import requests, json

R = requests.get('https://jsonplaceholder.typicode.com/posts/')
data = R.json()
m = [i for i in data if i['userId'] % 2 == 0]
print(json.dumps(m, indent=4))
```

```python
import json, requests

new_post = {"userId": 1,
            "title" : "Тестовый пост",
            "body" : "1 лаба по скриптовым"}

json_response = requests.post('https://jsonplaceholder.typicode.com/posts', json = new_post)
if json_response.ok:
    data = json_response.json()
    print(json.dumps(data, ensure_ascii = False, indent=3))
else:
    print(json_response.status_code)
```

```python
import json, requests

post_id = 1
url = f"https://jsonplaceholder.typicode.com/posts/{post_id}"

post_data_new = {
    "title": "Обновлённый пост",
    "body": "1 лаба по скриптовым",
    "id": post_id
}

response = requests.put(url, json=post_data_new)
post = response.json()

print(json.dumps(post, ensure_ascii=False, indent=4))
```
